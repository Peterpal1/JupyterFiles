```python
import matplotlib.pyplot as plt
import numpy as np


x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.xlabel('x- axis')

plt.ylabel('y- axis')

plt.title('x against y') 

plt.scatter(x,y,label ='stars',marker='*')
plt.legend()
plt.show()
```
